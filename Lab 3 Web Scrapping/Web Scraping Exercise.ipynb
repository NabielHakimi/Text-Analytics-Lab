### Question 1

**Print only quotes written by Albert Einstein, from all pages. Save the quotes in
“quotes_Einstein.csv”.**

```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python
# Function to scrape quotes from a page with Einstein filter
def scrape_einstein(soup, quotes):
    for quote in soup.find_all('div', class_='quote'):
        text = quote.find('span', class_='text').text
        author = quote.find('small', class_='author').text
        tags = ', '.join(tag.text for tag in quote.find_all('a', class_='tag'))
        
        # Only append if the author is Albert Einstein
        if author == "Albert Einstein":
            quotes.append({'Text': text, 'Author': author, 'Tags': tags})
```

```python
# Base URL and headers 
base_url = 'https://quotes.toscrape.com'
headers = {'User-Agent': 'Mozilla/5.0'}
einstein_quotes = []
```

```python
def scrape_all_pages(url):
    while url:
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.text, 'html.parser')
        scrape_einstein(soup, einstein_quotes)
        
        next_page = soup.find('li', class_='next')
        url = base_url + next_page.find('a')['href'] if next_page else None
```

```python
# Execution
scrape_all_pages(base_url)

# Save to quotes_Einstein.csv
with open('quotes_Einstein.csv', 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=['Text', 'Author', 'Tags'])
    writer.writeheader()
    writer.writerows(einstein_quotes)

print("Einstein quotes have been saved to quotes_Einstein.csv")
```

### Question 2

**Print only quotes containing the word “life” , from all pages. Save the quotes in
“life_quotes.csv”.**

```python
import requests
from bs4 import BeautifulSoup
import csv
```

```python
# Function to scrape quotes containing the word "life"
def scrape_life(soup, quotes):
    for quote in soup.find_all('div', class_='quote'):
        text = quote.find('span', class_='text').text
        author = quote.find('small', class_='author').text
        tags = ', '.join(tag.text for tag in quote.find_all('a', class_='tag'))
        
        # Check if "life" is in the quote text (case-insensitive check)
        if "life" in text.lower():
            quotes.append({'Text': text, 'Author': author, 'Tags': tags})
```

```python
base_url = 'https://quotes.toscrape.com'
headers = {'User-Agent': 'Mozilla/5.0'}
life_quotes = []
```

```python
def scrape_all_pages(url):
    while url:
        response = requests.get(url, headers=headers)
        soup = BeautifulSoup(response.text, 'html.parser')
        scrape_life(soup, life_quotes)
        
        next_page = soup.find('li', class_='next')
        url = base_url + next_page.find('a')['href'] if next_page else None
```

```python
# Execution
scrape_all_pages(base_url)

# Save to life_quotes.csv
with open('life_quotes.csv', 'w', newline='', encoding='utf-8') as csvfile:
    writer = csv.DictWriter(csvfile, fieldnames=['Text', 'Author', 'Tags'])
    writer.writeheader()
    writer.writerows(life_quotes)

print("Life quotes have been saved to life_quotes.csv")
```
